```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import cv2

print("NumPy version:", np.__version__)
print("Matplotlib version:", matplotlib.__version__)
print("OpenCV version:", cv2.__version__)
```

### 2.2 - Reading Images

```python
img_gray = cv2.imread('../database/chapter_2/lenna.png', cv2.IMREAD_GRAYSCALE)
print(img_gray)
```

```python
height, width = img_gray.shape
print(f"Image shape is: {width}x{height}")
```

### 2.3 - Displaying Images
Using Matplotlib

```python
plt.imshow(img_gray, cmap='gray')
plt.axis('off')
plt.show()
```

### 2.4 - Writing Images

```python
cv2.imwrite('../database/chapter_2/lenna_gray.png', img_gray)
```

* Setting parameters

### 2.7 - Converting between Classes 

```python
print("Default grayscale image data class:", img_gray.dtype)

img_gray_uint16 = img_gray.astype('uint16')

print("Image data class after conversion to uint16:", img_gray_uint16.dtype)
```

* Converting to binary

```python
_, img_binary = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY)

plt.imshow(img_binary, cmap='gray')
plt.axis('off')
plt.show()
```

### 2.8 - Array Indexing 

```python
# Vertical flip the image
img_gray_flipped = cv2.flip(img_gray, 0)
plt.imshow(img_gray_flipped, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Extract a section from the image
img_gray_section = img_gray[256:768, 256:768]
plt.imshow(img_gray_section, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Extract a section from the image
img_gray_subsample  = img_gray[::2, ::2]
height, width = img_gray_subsample.shape
print(f"Image shape is: {width}x{height}")
```

```python
# Plot a horizontal scan line through the middle of the image
plt.plot(img_gray[511, :])
plt.xlabel('Column Index')
plt.ylabel('Pixel Intensity')
plt.title('Horizontal Scan Line')
plt.show()
```

### 2.9 - Some Important Standard Arrays 

```python
M, N = 3, 3
# Generate an M x N matrix of zeros
zeros_matrix = np.zeros((M, N), dtype=float)
zeros_matrix
```

```python
# Generate an M x N matrix of ones
ones_matrix = np.ones((M, N), dtype=float)
ones_matrix
```

```python
# Generate an M x N boolean matrix of True values
true_matrix = np.full((M, N), True, dtype=bool)
true_matrix
```

```python
# Generate an M x N boolean matrix of False values
false_matrix = np.full((M, N), False, dtype=bool)
false_matrix
```

```python
# Generate an M x M identity matrix
identity_matrix = np.eye(M)
identity_matrix
```

```python
# Generate an M x N matrix of uniformly distributed random numbers between [0, 1)
random_matrix = np.random.rand(M, N)
random_matrix
```

```python
# Generate an M x N matrix of normally distributed random numbers with mean 0 and variance 1
gaussian_random_matrix = np.random.randn(M, N)

# Calculate the mean and variance
mean_value = np.mean(gaussian_random_matrix)
variance_value = np.var(gaussian_random_matrix)

print("Mean of the matrix:", mean_value)
print("Variance of the matrix:", variance_value)

gaussian_random_matrix
```
